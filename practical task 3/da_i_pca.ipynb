```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from pca import pca
```

```python
df = pd.read_csv('Ice.csv')
```

```python
df
```

```python
df_pca = pca(n_components = 0.95, normalize=True)

# Fit transform
pca_results = df_pca.fit_transform(df.iloc[:, :])
```

```python
print(pca_results)
```

```python
# Plot explained variance
fig, ax = df_pca.plot()
```

# So as we can see the most amount of information is in PC1 and a little bit less in PC2. They are the most sagnificant. 


```python
# Make biplot with the number of features
fig, ax = df_pca.biplot()#(n_feat=8)
```
